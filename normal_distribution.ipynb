```python
import numpy as np
from random import random
import matplotlib.pyplot as plt
from scipy.stats import normaltest
```

```python
def playRandomValue():
    return sum([random() for x in range(12)]) - 6

def showGraph(random_values):
    plt.hist(random_values, bins = 300)
    plt.show()
    
def get_mean(random_values):
    return sum(random_values) / len(random_values)

def get_variance(random_values):
    s = 0
    for i in range(len(random_values)):
        s += random_values[i] ** 2
    return s/len(random_values) - get_mean(random_values) ** 2
```

```python
random_values = [playRandomValue() for i in range(900)]
print(random_values)
```

```python
showGraph(random_values)
```

```python
mean = get_mean(random_values)
variance = get_variance(random_values)
```

```python
print(f'Среднее: {mean}\n')
print(f'Дисперсия: {variance}\n')
alpha = 0.05
print(f'Alpha: {alpha}\n')
k2, p = normaltest(random_values)
print(f'Probability: {p}\n')
if p < alpha:
    print("Нулевая гипотеза отвергается")
else:
    print("Нулевая гипотеза принята")
print()
```
